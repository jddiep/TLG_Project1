```python
%%writefile movie.py
import pandas as pd
import streamlit as st

names_link='/content/moviescsv.csv'
names_data=pd.read_csv(names_link)
data_movies = pd.read_csv(names_link)

sidebar = st.sidebar

sidebar.title("Movie App")
sidebar.write("Listado de películas")

st.title("Informacion sobre el Conjunto de Datos")
st.header("Información sobre Películas")

st.write("""Este es un ejemplo de una app para predecir
¡Esta app predice mis películas:
         """)

@st.cache_data

def load_data():
  datadirector = pd.read_csv(data_movies)
  return datadirector


@st.cache_data
def load_data_pordirector(director):
  datadirector = pd.read_csv(data_movies)
  filtered_data_pordirector = datadirector[datadirector['director'] == director]

  return filtered_data_pordirector

data= load_data
sidebar.header("Data Description")
selected_director = sidebar.selectbox("Select Director",data_movies['director'].unique())
sidebar.write("Selected Director:", selected_director)
btnFiltrarporDirector = sidebar.button('Filtrar por Director')
sidebar.markdown("___")


if (btnFiltrarporDirector):
  filtrarpordirector = load_data_pordirector(selected_director)
  st.dataframe(filtrarpordirector)

sidebar.header("Dataset")
agree = sidebar.checkbox("Show DataSet Overview ? ")
if agree:
 st.dataframe(data_movies)

@st.cache_data
def load_data_byname(name,nrows):
  data_movies = pd.read_csv(names_link, nrows=nrows)
  filtered_data_bymoviename = data_movies[data_movies['name'].str.contains(name)]

  return filtered_data_bymoviename

myname = sidebar.text_input('Titulo de película :')
mynamelower = myname.lower()
if (myname):
  data_load_state = st.text('Loading data...')
  filterbyname = load_data_byname(myname,500)
  count_row = filterbyname.shape[0]
  st.write(f"Total names : {count_row} ", mynamelower)
  data_load_state.text("Done! (using st.cache)")

  st.dataframe(filterbyname)


```
